```python
import json
import os
import matplotlib.pyplot as plt
from openqaoa.algorithms import QAOAResult
import numpy as np
```

## Upload json files

```python
dir_routes = ['/noiseless','/noisy','/zne']
dir_routes2 = ['./prob0','./prob1','./prob2','./prob3','./prob4']


results = {}

for dir_itr2,dir_route2 in enumerate(dir_routes2):
    results[dir_route2[2:]]={}
    for dir_itr,dir_route in enumerate(dir_routes):
        results[dir_route2[2:]][dir_route[1:]]={}
        for file_name in os.listdir(dir_route2+dir_route):
            if file_name.endswith(".json"):
                file_route = os.path.join(dir_route2+dir_route,file_name)
                with open(file_route, "r") as file:
                    result = json.load(file)
                    result = QAOAResult.from_dict(result)
                    results[dir_route2[2:]][dir_route[1:]][file_name[0:-5]] = result
```

```python
problem_keys = ['prob0','prob1','prob2','prob3','prob4']
environment_key = ['noiseless','noisy','zne']

def json_keys(prob_key,environment_key):
    if type(prob_key) ==str:
        return list(results[prob_key][environment_key].keys())
    else:
        return list(prob_key[environment_key].keys())

def json_key(prob_key,environment_key):
    if type(prob_key) ==str:
        return list(results[prob_key][environment_key].keys())[0]
    else:
        return list(prob_key[environment_key].keys())[0]


print(json_key('prob1','noiseless'),'\n')
print(json_key(results['prob2'],'noisy'),'\n')
print(json_keys('prob0','zne'),'\n')
print(json_keys(results['prob3'],'zne'),'\n')
```

```python
def calculate_metrics(optimal_cost,problem):
    noiseless_cost = abs(problem['noiseless'][json_key(problem,'noiseless')].optimized['cost'])
    noisy_cost = abs(problem['noisy'][json_key(problem,'noisy')].optimized['cost'])

    noiseless_aprox = approximation_ratio(noiseless_cost,optimal_cost)
    noisy_aprox = approximation_ratio(noisy_cost,optimal_cost)
    zne_aproxs = {}

    noiseless_fev =problem['noiseless'][json_key(problem,'noiseless')].evals['number_of_evals']
    noisy_fev =problem['noisy'][json_key(problem,'noisy')].evals['number_of_evals']
    zne_fevs = {}

    for zne_key in json_keys(problem,'zne'):   
        zne_aproxs[zne_key] = approximation_ratio(abs(problem['zne'][zne_key].optimized['cost']),optimal_cost)
        zne_fevs[zne_key] = problem['zne'][zne_key].evals['number_of_evals']

    categories = ['noiseless','noisy','conf. 0','conf. 1','conf. 2','conf. 3','conf. 4','conf. 5','conf. 6','conf. 7']
    aproxs = [noiseless_aprox,noisy_aprox]
    fevs = [noiseless_fev,noisy_fev]
    for zne_key in json_keys(problem,'zne'):   
        #zne_label = zne_key[zne_key.find("conf"):]
        #categories.append(zne_label)
        aproxs.append(zne_aproxs[zne_key])
        fevs.append(zne_fevs[zne_key])
    return categories,aproxs,fevs
```

```python
optimal_costs = [abs(-9),abs(-16),abs(-12),abs(-12),abs(-12)] # took from maxcut_qaoa_zne notebook, and calculate it absolute value

ancho = 0.19
#x = np.arange(len(categories))
fig, ax = plt.subplots(figsize=(8, 6))
fig2, ax2 = plt.subplots(figsize=(7.2, 6))

colors = ['#AF47D2', '#FF8F00', '#6499E9']


for itr,optimal_cost in enumerate(optimal_costs[:3]):
    categories,aproxs,fevs = calculate_metrics(optimal_cost,results[problem_keys[itr]])

    x = np.arange(len(categories))
    positions = [x - ancho,x,x + ancho]
    
    ax.bar(positions[itr], aproxs, ancho, label='instance '+str(itr),color=colors[itr])
    ax2.bar(positions[itr], fevs, ancho, label='instance '+str(itr),color=colors[itr])

ax.set_xlabel('environment',fontsize=14)
ax.set_ylabel('approx. ratio',fontsize=14)
ax.set_title('Approximation ratio for each environment',fontsize=16)
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=45, ha='right')
ax.legend()

ax2.set_xlabel('environment',fontsize=14)
ax2.set_ylabel('function evaluations',fontsize=14)
ax2.set_title('Function evaluations for each environment',fontsize=16)
ax2.set_xticks(x)
ax2.set_xticklabels(categories, rotation=45, ha='right')
ax2.legend()

#ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')


# Mostrar el grfico
plt.tight_layout()

fig.savefig('imgs/approxratio_comparison_pro0prob1prob2.pdf')
fig2.savefig('imgs/fevs_comparison_pro0prob1prob2.pdf')


plt.show()
```
